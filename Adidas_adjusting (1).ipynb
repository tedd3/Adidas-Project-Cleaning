```python
import pandas as pd
```

```python
data = pd.read_excel('Book1.xlsx')
```

```python
data
```

```python
data.info()
```

```python
# Set datatype cho từng cột
data['Retailer'] = data['Retailer'].astype(str)
data['Retailer ID'] = data['Retailer ID'].astype(str)
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
data['Region'] = data['Region'].astype(str)
data['State'] = data['State'].astype(str)
data['City'] = data['City'].astype(str)
data['Product'] = data['Product'].astype(str)
data['Price per Unit'] = data['Price per Unit'].astype(float)
data['Units Sold'] = data['Units Sold'].astype(int)
data['Total Sales'] = data['Total Sales'].astype(float)
data['Operating Profit'] = data['Operating Profit'].astype(float)
data['Operating Margin'] = data['Operating Margin'].astype(float)
data['Sales Method'] = data['Sales Method'].astype(str)
```

```python
data
```

```python
# Tạo cột "Quarter" và "Year" từ cột "Date"
data['Quarter'] = pd.to_datetime(data['Invoice Date']).dt.quarter
data['Year'] = pd.to_datetime(data['Invoice Date']).dt.year
```

```python
# Tách dữ liệu từ cột "Product" thành "Gender" và "Product_type"
data['Gender'] = data['Product'].str.split("'s").str[0]
data['Product_type'] = data['Product'].str.split("'s").str[1]
```

```python
# Tính trung bình của cột "Operating Margin"
mean_operating_margin = data['Operating Margin'].mean()
# Hàm để kiểm tra điều kiện và tạo giá trị cho cột "Profit_efficiency"
def profit_efficiency(row):
    if row['Operating Margin'] > mean_operating_margin:
        return'Efficient'
    else:
        return 'Inefficient'

# Áp dụng hàm "profit_efficiency" vào DataFrame
data['Profit_efficiency'] = data.apply(profit_efficiency, axis=1)
```

```python
data.dropna()
```

```python
data.drop(columns = 'Unnamed: 14', inplace=True)
```
